# Simple Response Time Measurements from Production Eric
This document is a Jupyter Notebook containing a simple analysis of some performance measurements from production Eric.

The details used here were recorded using a Selenium script which saved results to an Excel file, eric_data.xlsx.

The selenium script can be found here https://github.com/alanmaydwell/eric_selenium 

Items covered here:

- Application launch time from portal
- Time taken to complete a search
- Time taken to select a report (for each of four report types)
- Time taken to view a report (for each of four report types)

***
*Start technical setup*

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Access Excel file
##%cd ~/gits/eric_selenium
filename = "eric_data.xlsx"
xl = pd.ExcelFile(filename)
```

```python
# Read data from the Results tab of the Excel
df_all = xl.parse("Results", skiprows=4)
# Set dates to be dates as dates
##df_all.Date = pd.to_datetime(df_all.Date, dayfirst=True)
```

```python
# Extract some sub-sets from the data
df_initial = df_all[["Date", "Launch Time", "Search Time"]]
df_report1 = df_all[["Date", "Report 1 Select Time", "Report 1 View Time"]]
df_report2 = df_all[["Date", "Report 2 Select Time", "Report 2 View Time"]]
df_report3 = df_all[["Date", "Report 3 Select Time", "Report 3 View Time"]]
df_report4 = df_all[["Date", "Report 4 Select Time", "Report 4 View Time"]]
```

```python
# Graph plotting function
def plot_df(df, title=""):
    """plot a pandas dataframe"""
    df.plot(x='Date', figsize=(10,6))
    plt.xticks(rotation=90)
    plt.ylim(ymin=0)
    plt.ylabel("Time(s)")
    #plt.legend(loc='best')
    plt.legend(loc=3)
    plt.title(title, fontsize=16)
```

*End of technical setup*
***

## Application Launch and Search
1. Time taken for Eric application to open after clicking link in portal. (login to portal itself not included)
2. Time taken to complete a search within Eric

```python
plot_df(df_initial, "Application Launch and Search")
df_initial.describe()
```

## Report 1 "Civil Financial Statement"


```python
plot_df(df_report1, "Civil financial statement")
df_report1.describe()
```

## Report 2 "Criminal Financial Statement"

```python
plot_df(df_report2, "Criminal financial statement")
df_report2.describe()
```

## Report 3 "Family Mediation Financial Statement"

```python
plot_df(df_report3, "Family Mediation Financial Statement")
df_report3.describe()
```

## Report 4 "Financial Statement Summary"

```python
plot_df(df_report4, "Financial Statement Summary")
df_report4.describe()
```
